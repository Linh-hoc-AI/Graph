```python
class Graph:
    def __init__(self):
        self.adjacency_list = {}
        self.adjacency_matrix = {}

    def add_edge(self, u, v, weight=None):
        # Thêm cạnh vào danh sách kề
        if u not in self.adjacency_list:
            self.adjacency_list[u] = []
        self.adjacency_list[u].append((v, weight))

        # Thêm cạnh vào ma trận kề
        if u not in self.adjacency_matrix:
            self.adjacency_matrix[u] = {}
        self.adjacency_matrix[u][v] = weight

    def display_adjacency_list(self):
        print("Adjacency List:")
        for vertex, neighbors in self.adjacency_list.items():
            print(f"{vertex}: {neighbors}")

    def display_adjacency_matrix(self):
        print("\nAdjacency Matrix:")
        for vertex, neighbors in self.adjacency_matrix.items():
            row = []
            for v in self.adjacency_list.keys():
                row.append(neighbors.get(v, 0))
            print(row)

# Chuyển đổi danh sách cạnh thành danh sách kề và ma trận kề
def convert_edges_to_graph_representation(edges):
    graph = Graph()
    for edge in edges:
        u, v, weight = edge
        graph.add_edge(u, v, weight)
    return graph

# Ví dụ danh sách cạnh
edges = [('A', 'B', 2), ('A', 'C', 1), ('B', 'C', 3), ('B', 'D', 4)]

# Chuyển đổi thành đồ thị và hiển thị danh sách kề và ma trận kề
graph = convert_edges_to_graph_representation(edges)
graph.display_adjacency_list()
graph.display_adjacency_matrix()

# DFS algorithm
# Định nghĩa đồ thị bằng danh sách kề
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F', 'G'],
    'D': [],
    'E': ['H'],
    'F': [],
    'G': [],
    'H': []
}

# Hàm DFS
def dfs(graph, start, visited):
    # Nếu đỉnh hiện tại chưa được ghé thăm, in và đánh dấu đã ghé thăm
    if start not in visited:
        print(start, end=' ')
        visited.add(start)
        # Duyệt qua tất cả các đỉnh kề của đỉnh hiện tại và gọi đệ quy
        for neighbor in graph[start]:
            dfs(graph, neighbor, visited)

# Gọi hàm DFS với đỉnh bắt đầu là 'A'
visited = set()  # Sử dụng tập hợp để theo dõi các đỉnh đã ghé thăm
dfs(graph, 'A', visited)

```
